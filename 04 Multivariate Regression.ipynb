## Correlation
## $$ \rho _{XY} = corr(X, Y) $$
## $$ -1.0 \leq \rho _{XY} \leq + 1.0 $$

```python
 data['PRICE'].corr(data['RM'])
```

```python
# Calculate the correlation between property prices and the pupil theaacher ratio
data['PRICE'].corr(data['PTRATIO'])
```

```python
data.corr()  # Pearson Correlation Coefficients
```

```python
mask = np.zeros_like(data.corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices] = True
mask
```

```python
plt.figure(figsize =(16, 10))
sns.heatmap(data.corr(), mask=mask, annot=True, annot_kws={'size': 14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
# Picture the relationship between pollution and distance in your head 
# Then create a scatter plot between DIS and NOX

nox_dis_corr = round(data['NOX'].corr(data['DIS']), 3)

plt.figure(figsize=(16, 10))
plt.scatter(x=data['DIS'], y=data['NOX'], alpha=0.6, s=100, color='indigo')

plt.title(f'DIS vs NOX (Correlation {nox_dis_corr})', fontsize=14)
plt.xlabel('DIS - Distance from employment', fontsize=14)
plt.ylabel('NOX - Nitric Oxide Pollution', fontsize=14)
plt.show()
```

```python
sns.set()
sns.set_context('talk')
sns.set_style('whitegrid')
sns.jointplot(x=data['DIS'], y=data['NOX'], height=7, color='indigo', joint_kws={'alpha':0.5})
plt.show()   
```

```python
sns.set()
sns.set_context('talk')
sns.set_style('whitegrid')
sns.jointplot(x=data['DIS'], y=data['NOX'], kind='hex', height=7, color='blue')
plt.show()   
```
